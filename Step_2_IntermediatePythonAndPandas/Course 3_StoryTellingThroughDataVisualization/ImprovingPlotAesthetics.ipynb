#### **Title**

StoryTelling Through Data Visualization

##### **DataSet Details**
[The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp)
releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

##### **Data Credits**
Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled [here](www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv).

Randal compiled this data set to explore the gender gap in STEM fields, which stands for science, technology, engineering, and mathematics. This gap is reported on often in the news and not everyone agrees that there is a gap.

##### **Objective**
In this analysis we'll explore how we can communicate the nuanced narrative of gender gap using effective data visualization.
If we want the gender gap to be apparent and emphasized in the plot, we need a visual analogy to the difference in the percentages between the genders.

##### **Data Introduction**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('../../datasets/percent-bachelors-degrees-women-usa.csv')

print('[COLUMN DETAILS]')
print(women_degrees.iloc[0], '\n')
```

#### **Visualizing historical percentage of Biology degrees awarded to women**

```python
# 3. Introduction To The Data
plt.plot(women_degrees['Year'], women_degrees['Biology'])
plt.show()
```

From the above plot, we can tell that Biology degrees increased steadily from 1970 and peaked in the early 2000's. Also, the percentage has stayed above 50% since around 1987. 

If we visualize the trend of biology degrees awarded to men on the same plot, a viewer can observe the space between the lines for each gender.

#### **Visualize the trend of Biology degrees awarded to men**
We can calculate the percentages of Biology degrees awarded to men by subtracting each value in the Biology column from 100.

```python
# 4. Visualizing The Gender Gap
plt.plot(women_degrees['Year'], women_degrees['Biology'], c='red', label='Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='blue', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

The chart containing both line charts tells a more complete story than the one containing just the line chart that visualized just the women percentages. This plot instead tells the story of two distinct periods. 

The first period, from 1970 to around 1987, women were a minority when it came to majoring in Biology while in the second period, from around 1987 to around 2012, women became a majority.
One can observe the point where women overtook men where the lines intersect.

##### **Improving Plot Aesthetics**

```python
# 6. Hiding Tick Marks

fig, ax = plt.subplots()


ax.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', 
        label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green',
        label='Men')

ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc='upper right')
ax.tick_params(bottom=False,left=False) 
plt.show()
```

```python
# 7. Hiding Spines

fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green', label='Men')
ax.tick_params(bottom=False,left=False) 

for key,spine in ax.spines.items():
    spine.set_visible(False)
    
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc='upper right')
plt.show()
```

#### **Comparing Gender Gap Across Degree Categories**
Generating line charts for four STEM degree categories on a grid to encourage comparison.

```python
# 8.Comparing gender gap across degree categories

categories = ['Biology', 'Computer Science', 'Engineering', 
              'Math and Statistics']
fig = plt.figure(figsize=(12, 12))


for p in range(0, 4):
    ax = fig.add_subplot(2, 2, p+1)
    ax.plot(women_degrees['Year'], women_degrees[categories[p]],
            c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[categories[p]],
            c='green', label='Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.set_title(categories[p])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    
plt.legend(loc='lower right')    
plt.show()
```

We can conclude that the gender gap in Computer Science and Engineering is large, while the gap in Biology and Math and Statistics is quite small. In addition, the first two degree categories are dominated by men while the latter degree categories are much more balanced. 

#### **Color, Layout, and Annotations**


##### **Setting Line Color Using RGB**
##### **Setting Line Width**

```python
categories = ['Biology', 'Computer Science', 'Engineering', 
              'Math and Statistics']

dark_blue = (0/255, 107/255, 164/255)
orange = (255/255, 128/255, 14/255)
fig = plt.figure(figsize=(12, 12))

for p in range(0,4):
    ax = fig.add_subplot(2,2,p+1)
    # The color for each line is assigned here.
    ax.plot(women_degrees['Year'], women_degrees[categories[p]], 
            c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[categories[p]], 
            c=orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(categories[p])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

##### **Improve Layout And Ordering**
To make the viewing experience more coherent
* Use layout of a single row with multiple columns
* Order the plots in decreasing order of initial gender gap

```python
, # 5. Improve the layout and ordering

# stem_categories list has been populated with six STEM degree categories, 
# ordering them by decreasing ending gender gap.
stem_categories = ['Engineering', 'Computer Science', 'Psychology', 'Biology',
                   'Physical Sciences', 'Math and Statistics']
dark_blue = (0/255, 107/255, 164/255)
orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize=(18, 3))

for p in range(0,6):
    ax = fig.add_subplot(1,6,p+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_categories[p]], 
            c=dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_categories[p]], 
            c=orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_categories[p])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()                          
```

The leftmost plot has the largest gender gap in 1968 while the rightmost plot has the smallest gender gap in 1968. 

##### **Replacing The Legend With Annotations**
We replace the legend entirely by annotating the lines directly with the corresponding genders. Even the position of the text annotations have meaning. In both plots, the annotation for Men is positioned above the orange line while the annotation for women is positioned below the dark blue line. This position subtly suggests that **Men** are a majority for the degree categories that the line charts are representing **(Engineering & Math and Statistics)** and **Women** are a minority for those degree categories. 

```python
# 7. Annotating in Matplotlib
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize=(22, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()  
```
