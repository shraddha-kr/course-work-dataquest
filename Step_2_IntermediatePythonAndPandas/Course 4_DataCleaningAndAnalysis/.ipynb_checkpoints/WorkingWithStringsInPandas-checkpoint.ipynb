```python
import pandas as pd

happiness2015 = pd.read_csv('./dataset/World_Happiness_2015.csv')
world_dev = pd.read_csv("./dataset/World_dev.csv")
world_dev.head()
```

##### Merging two datasets

```python
col_renaming = {'SourceOfMostRecentIncomeAndExpenditureData': 'IESurvey'}
merged = pd.merge(left=happiness2015, right=world_dev, how='left', left_on='Country', right_on='ShortName')
merged = merged.rename(col_renaming, axis=1)

print(merged.info())
merged.head()
```

#### WORKING WITH STRINGS IN PANDAS 

##### Using Apply to Transform Strings

```python
def extract_last_word(element):
    return str(element).split()[-1]

merged['Currency Apply'] = merged['CurrencyUnit'].apply(extract_last_word)
merged['Currency Apply']
```

We should use built-in vectorized methods(if they exist) instead of the *Series.apply()* method for performance reasons. Instead we could have split each element in the column into a list of strings with the Series.str.split() method, the vectorized equivalent of Python's *string.spit()*
![title](./img/VectorizedString_1.png)

##### Vectorized String Methods Overview
Using vectorized string methods results:
1. Better performance
2. Code that is easier to read and write
3. Automatically excludes missing values

Below are some common vectorized string methods, but you can find the full list [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html)
![title](./img/VectorizedString_2.png)

We access these vectorized string methods by adding a str between the Series name and method name:
![title](./img/VectorizedString_3.png)
The str attribute indicates that each object in the Series should be treated as a string, without us having to explicitly change the type to a string like we did when using the apply method.

##### Slicing & Method Chaining
We can also slice each element in the Series to extract characters, but we did still need to use the str attribute. 

```python
merged['CurrencyUnit'].str[0:5]
```

Vectorized string methods can be chained. Suppose we needed to split each element in the columnn into a list of strings using the *Series.str.split()* method and capitalize the letters using the 'Series.str.upper()'.

```python
merged['CurrencyUnit'].str.upper().str.split()
```

```python
merged['Currency Vectorized'] = merged['CurrencyUnit'].str.split().str.get(-1).str.upper()
merged['Currency Vectorized'].head()
```

##### Exploring Missing Values with Vectorized String Methods
Missing values shouldn't be treated as strings. **NaN** gets treated as a string,they should instead have been excluded from the calculation.

```python
def compute_lengths(element):
    if pd.isnull(element):
        pass
    else:
        return len(str(element))
    
lengths_apply = merged['CurrencyUnit'].apply(compute_lengths)
print(lengths_apply.head())
merged['CurrencyUnit'].head()
```

```python
lengths = merged['CurrencyUnit'].str.len()
value_counts = lengths.value_counts(dropna=False)
value_counts
```

##### Finding Specific Words in Strings

Suppose we needed to parse the elements of a Series to find a string or substring that doesn't appear in the same position in each string. We can handle problems like this with **regular expressions**, or **regex**. A regular expression is a sequence of characters that describe a search pattern, used to match characters in a string.
![title](./img/Regex_1.png)

In pandas, regular expression is integrated with vectorized string methods to make finding and extracting patterns of characters easier. However the rules for creating regular expressions can be quite complex. 

```python
pattern = r"[Nn]ational accounts"
national_accounts = merged['SpecialNotes'].str.contains(pattern)
print(national_accounts.head())
merged['SpecialNotes'].head()
```

We change the **NaN** values to False
![title](./img/Regex_2.png)

```python
pattern = r"[Nn]ational accounts"
national_accounts = merged['SpecialNotes'].str.contains(r"[Nn]ational accounts", na=False)
merged_national_accounts = merged[national_accounts]

print(national_accounts)
merged_national_accounts.head()
```

The characters in a year follow a specific pattern
![title](./img/Regex_3.png)

```python
pattern = r"([1-2][0-9]{3})"
years = merged['SpecialNotes'].str.extract(pattern, expand=True)
years
```


When we used the *Series.str.extract* method, we enclosed our regular expression in parentheses. The parentheses indicate that only the character pattern matched should be extracted and returned in a series, this is called the capturing group.
If the capturing group doesn't exist in a row(or there is no match) the value in that row is set to NaN.
![title](./img/Regex_6.png)
![title](./img/regex_7.png)

We can also return the results as a dataframe by changing the **expand** parameter to True.
The Series.str.extract() method will only extract the first match of the pattern. If we wanted to extract all of the matches, we can use the **Series.str.extractall()** method.

![title](./img/Regex_4.png)

##### Extracting All Matches of a Pattern from a Series

We will use the same regular expression from the last screen to extract all the years, we will use a named capturing group. A named capturing group means that we can refer to the group by the specified name instead of just a number. Following syntax can be used to add a name (?P<Column_Name>)

```python
# Setting the Country column as the index
merged = merged.set_index('Country')

pattern = r"(?P<Years>[1-2][0-9]{3})"
years = merged['SpecialNotes'].str.extractall(pattern)
years
```

![title](./img/Regex_8.png)

```python
pattern = r"(?P<Years>[1-2][0-9]{3})"
years = merged['IESurvey'].str.extractall(pattern)
value_counts = years['Years'].value_counts()
value_counts.sort_index()
```

##### Extracting More Than One Group of Patterns from a Series
When we tried to extract all of the years from the column using the extractall method, we were unsuccessful because some of our years had the following format:

![title](./img/Regex_9.png)

```python
pattern =  r"(?P<First_Year>[1-2][0-9]{3})(/)?(?P<Second_Year>[0-9]{2})?"
years = merged['IESurvey'].str.extractall(pattern)
years.sort_values('Second_Year')
```

Note that we also added a question mark, ?, after each of the two new groups to indicate that a match for those groups is optional. This allows us to extract years listed in the yyyy format AND the yyyy/yy format at once.

```python
pattern = r"(?P<First_Year>[1-2][0-9]{3})/?(?P<Second_Year>[0-9]{2})?"
years = merged['IESurvey'].str.extractall(pattern)
first_two_year = years['First_Year'].str[0:2]
years['Second_Year'] = first_two_year + years['Second_Year']
years
```

1. If part of the regex isn't grouped using parantheses, (), it won't be extracted.
2. When we add a string to a column using the plus sign, +, pandas will add that string to every value in the column. Note that the strings will be added together without any spaces.


##### Clean a String Column, Aggregate the Data, and Plot the Results

```python
pv_incomes = merged.pivot_table(values='Happiness Score', index='IncomeGroup')
print(pv_incomes)
pv_incomes.plot(kind='bar', rot=30, ylim=(0,10))
```

```python
import matplotlib.pyplot as plt

merged['IncomeGroup'] = merged['IncomeGroup'].str.replace(' income', '').str.replace(':', '').str.upper()
pv_incomes = merged.pivot_table(values='Happiness Score', index='IncomeGroup')
pv_incomes.plot(kind='bar', rot=30, ylim=(0,10))
plt.show()
```
