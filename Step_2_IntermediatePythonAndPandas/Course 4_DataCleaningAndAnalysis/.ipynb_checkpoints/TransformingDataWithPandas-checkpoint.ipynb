#### OBJECTIVE
Following are the factors which contribute to the happiness score:
* Economy (GDP per capita)
* Family
* Health (Life Expectancy)
* Freedom

**Which of the factors above contribute the most to the happiness score?**

#### DATA INTRODUCTION

We shall focus on the 2015 report.
Below are descriptions for some of the columns that we will work on,

| Column      | Definition |
| :---------- | :--------- |
| **Country**  |Name of the country.|
| **Region**     | Name of the region the country belongs to.|
| **Happiness Rank**  |The rank of the country, as determined by its happiness score.|

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

happiness2015 = pd.read_csv('./dataset/World_Happiness_2015.csv')
first_5 = happiness2015.head()
print(happiness2015.info())
first_5
```

#### TRANSFORMING DATA
##### Re-naming Columns

```python
mapping = {'Economy (GDP per Capita)': 'Economy', 
           'Health (Life Expectancy)': 'Health', 
           'Trust (Government Corruption)': 'Trust' }
happiness2015 = happiness2015.rename(mapping, axis = 1)
happiness2015.info()
```

##### Apply a Function Element-Wise Using the Map and Apply Methods
![title](./img/TransformData_1.png)
![title](./img/TransformData_2.png)

Each of the factor columns consists of numbers. Each number represents the extent to which each factor contributes to the happiness score. Since it is challenging to analyze all of these numbers across multiple columns. We can first convert these numbers to categories that indicate whether the factor has a high impact on the happiness score or a low impact.

Both methods above apply a function element-wise to a column. When we say element-wise, we mean that we pass the function one value in the series at a time and it performs some kind of transformation. 

Note that these methods both take a function as a parameter. Because we are using the function as a parameter, we pass it into the function without the parentheses.

```python
def label(element):
    if element > 1:
        return 'High'
    else:
        return 'Low'

economy_impact_map = happiness2015['Economy'].map(label)
economy_impact_apply = happiness2015['Economy'].apply(label)
equal = economy_impact_map.equals(economy_impact_apply)
equal
```

We applied a function to the Economy column using the Series.map() and Series.apply() methods and confirmed that both methods produce the same results. These methods don't modify the original series. Inorder to work with the new series in the original dataframe, we must either assign the results back to the original column or create a new column.
![title](./img/TransformData_3.png)

To create the Economy Impact column, map() and apply() iterate through the Economy column and pass each value into the label function. The function evaluates which range the value belongs to and assigns the corresponding value to the element in the new column.

```python
def label(element):
    if elementment > 1:
        return 'High'
    else:
        return 'Low'

def label(element, x):
    if element > x:
        return 'High'
    else:
        return 'Low'

economy_impact_apply = happiness2015['Economy'].apply(label, x=0.8)
economy_impact_apply
```

We can only use the Series.apply() method to apply a function with additional arguments element-wise. The Series.map() method will return an error.

 *TypeError: map() got an unexpected keyword argument 'x'*


##### Apply a Function Element-Wise to Multiple Columns Using the ApplyMap Method
Pandas already has a method that can apply function element-wise to multiple columns at once - the **DataFrame.applymap()** method. We need to pass the function name into the df.applymap() method without parentheses.

```python
def label(element):
    if element > 1:
        return 'High'
    else:
        return 'Low'
    
economy_apply = happiness2015['Economy'].apply(label)
factors = ['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity']
factors_impact = happiness2015[factors].applymap(label)
factors_impact
```

We can see from the results that, **Economy** and **Family** columns had a high impact on the happiness scores of the first five countries

##### Apply Functions Along an Axis Using the Apply Method
Use the apply() method on a dataframe, but the **DataFrame.apply()** method has different capabilities. Instead of applying functions element-wise, the df.apply() method applies functions along an axis, either column-wise or row-wise. When we create a function to use with df.apply(), we set it up to accept a series, most commonly a column. 

```python
factors_impact.apply(pd.value_counts)
```

![title](./img/TransformData_5.png)
When we use the df.apply() method to calculate the number of 'High' and 'Low' values in each column of the result, we can easily see that the Family and Economy columns contain the most High values.

We could use the df.apply() method to transform multiple columns, only because the pd.value_counts function operates on a Series. If we tried to use the df.apply() method to apply a function that works element-wise to multiple columns, we will get an error.

*ValueError: ('The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().', u'occurred at index Economy')*

```python
def v_counts(col):
    num = col.value_counts()
    den = col.size
    return num/den

v_counts_pct = factors_impact.apply(v_counts)
v_counts_pct
```

The result is a dataframe containing the percentage of 'High' and 'Low' values in each column.

We should only use the apply() method when a vectorized function does not exist. Recall that pandas uses vectorization, the process of applying operations to whole series at once, to optimize performance. When we use the apply() method, we're actually looping through rows, so a vectorized method can perform an equivalent task faster than the apply() method. 

##### Calculate the sum of the factor columns in each row
The sum of the six factors and the 'Dystopia Residual' column equals the happiness score

```python
#Calculate the sum of the factor columns in each row.
happiness2015['Factors Sum'] = happiness2015[['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']].sum(axis=1)

#Display the first five rows of the result and the Happiness Score column.
happiness2015[['Happiness Score', 'Factors Sum']].head()
```

Because the sum of the seven columns equal the happiness score, we can convert them to percentages and analyze them as proportions of the happiness score.

```python
factors = ['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']

def percentages(col):    
    # (Economy/Total happiness score) * 100
    div = col/happiness2015['Happiness Score']     
    return div * 100

factor_percentages = happiness2015[factors].apply(percentages)
factor_percentages
```

##### Reshaping Data with the Melt Function

![title](./img/Melt_2.png)

We will use melt function to reshape the above subset so that the values for Economy, Family and Health reside in the same column.

```python
factors = ['Country', 'Happiness Score','Economy', 'Family', 'Health']
happy_two = happiness2015[factors].head(2)
happy_two
```

```python
pd.melt(happy_two, id_vars=['Country'], value_vars=['Economy', 'Family', 'Health'])
```

![title](./img/Melt_1.png)

```python
main_cols = ['Country', 'Region', 'Happiness Rank', 'Happiness Score']
factors = ['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual']

melt = pd.melt(happiness2015, id_vars = main_cols, value_vars = factors)
melt['Percentage'] = round(melt['value']/melt['Happiness Score'] * 100, 2)
melt
```

The melt function was used to reshape our data so that we could use vectorized operations to convert the value column into percentages. It moved the values in the seven columns, to the same column, which helps in transforming them all at once.

Data in a format that makes it easier to aggregate, this format is referred to as **tidy** data


#### TRANSFORM DATA AND CREATE A VISUALIZATION

```python
melt = pd.melt(happiness2015, id_vars = ['Country', 'Region', 'Happiness Rank', 'Happiness Score'], value_vars= ['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia Residual'])
melt['Percentage'] = round(melt['value']/melt['Happiness Score'] * 100, 2)
pv_melt = melt.pivot_table(index='variable', values='value')
pv_melt.plot(kind='pie', y='value', legend=False)
```
